# CairoMakieに関する描画のノートブック

## 概要
本ノートは，研究ノートブックのテンプレートである．

```julia
# Color definitions
color = Dict(
    "red8" => (193,39,45) ./ 255,
    "orange8" => (247,147,30) ./ 255,
    "yellow8" => (238,204,223) ./ 255,
    "green8" => (55,188,0) ./ 255,
    "cyan8" => (0,169,157) ./ 255,
    "blue8" => (0,113,188) ./ 255,
    "navy8" => (0,0,178) ./ 255,
    "purple8" => (104,2,193) ./ 255,
)
```

```julia
using CairoMakie

# Data Set
seconds = 0:0.1:2
measurements = [8.2, 8.4, 6.3, 9.5, 9.1, 10.5, 8.6, 8.2, 10.5, 8.5, 7.2,
        8.8, 9.7, 10.8, 12.5, 11.6, 12.1, 12.1, 15.1, 14.7, 13.1]

# Plotting
f = Figure()
ax = Axis(f[1, 1],
    title = "Experimental data and exponential fit",
    xlabel = "Time (seconds)",
    ylabel = "Value",
)
scatter!(
    ax,
    seconds,
    measurements,
    color = RGBf(color["red8"]...),
    label = "Measurements"
)
lines!(
    ax,
    seconds,
    exp.(seconds) .+ 7,
    color = RGBf(color["orange8"]...),
    linestyle = :dash,
    label = L"f(x) = \exp(x) + 7",
)
axislegend(position = :rb)

save("../paper/AIAA_template/img/example.png", f; px_per_unit=4)
f
```
